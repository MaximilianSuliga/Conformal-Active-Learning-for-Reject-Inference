```python
df_list=[Small, German, Deloitte, Large]
```

```python
# Delete columns with more than 30% missing values
for df in df_list:
    df = df[df.columns[df.isnull().mean() < 0.3]]
```

```python
for df in df_list:
    df.drop_duplicates()
```

```python
# Fill missing values
for df in df_list:


    for feature in df.select_dtypes(exclude="object_").columns:
        df[feature].fillna(df[feature].mean(), inplace=True)

    # Fill missing values
    for feature in df.select_dtypes("object").columns:
        most_frequent_category = df[feature].mode().values[0]
        df[feature].fillna(most_frequent_category, inplace=True)
```

```python
# Drop features with only one unique value
for df in df_list:
    for col in df.columns:
        if df[col].nunique() == 1:
            df.drop(col, axis=1, inplace=True)
```

```python
# Simulate datasets
simulations=[2000000]
simulated_dfs = simulation_demonstration(df_list, simulations, return_dfs=True)
```

```python
simulated_dfs = [df[0] for df in simulated_dfs]
```

```python
# Create starting round, original datasets, MC datasets
dfs_start=[]
dfs_al=[]
dfs_al_mc=[]

for df, sim in zip(df_list, simulated_dfs):
    start, al, al_mc = al_splitting(df,sim)
    dfs_start.append(start)
    dfs_al.append(al)
    dfs_al_mc.append(al_mc)
```

```python
# Save as csv
for i, df in enumerate(dfs_start):
    df.to_csv(f'start_{i}.csv', index=False)
    
for i, df in enumerate(dfs_al):
    df.to_csv(f'AL_original_{i}.csv', index=False)
    
for i, df in enumerate(dfs_al_mc):
    df.to_csv(f'AL_MC_{i}.csv', index=False)    
```
