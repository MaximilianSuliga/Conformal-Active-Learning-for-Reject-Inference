```python
import numpy as np
import pandas as pd
import pickle
from sklearn.model_selection import train_test_split

pd.set_option('display.max_columns', None)

# Read data
rejected = pd.read_csv("rejected_2007_to_2018Q4.csv")
accepted = pd.read_csv("accepted_2007_to_2018Q4.csv")
```

```python
rejected.describe(include = 'all')
```

```python
accepted.describe(include = 'all')
```

```python
# Initial amount of accepted loans
in_size_acc=len(accepted)
in_size_acc
```

```python
# Initial amount of rejected loans
in_size_rej=len(rejected)
in_size_rej
```

```python
# Rejection rate
rejection_rate = in_size_rej / in_size_acc
rejection_rate
```

```python
# Check for missing values among rejections
rejected.isnull().sum()
```

```python
# Check if there are missing values for the loan status
accepted['loan_status'].isnull().sum()
```

```python
# Loan status categories
accepted['loan_status'].unique()
```

```python
# Delete rows with unavailable loan status
accepted = accepted.dropna(subset=['loan_status'])
```

```python
# Neither good nor bad loans
excluded_values = ['Current', 'Late (31-120 days)', 'In Grace Period', 'Late (16-30 days)']

# Exclude them to have only good or bad loans in the dataset
accepted = accepted.loc[~accepted['loan_status'].isin(excluded_values)]
```

```python
# Create BAD variable
accepted['BAD']=1
```

```python
# Create a list with those characteristics, that indicate a good loan
still_good = ['Fully Paid', 'Does not meet the credit policy. Status:Fully Paid']
```

```python
# Set the BAD variable to 1 for every loan that is indicated as good by its status
accepted.loc[accepted['loan_status'].isin(still_good), 'BAD'] = 0
```

```python
accepted
```

```python
# Create a BAD variable for the dataset of rejections
rejected['BAD']=1
```

```python
rejected.describe(include='all')
```

```python
accepted.describe(include='all')
```

```python
# Create a dictionary to indicate how to rename which columns
rename_dict = {'loan_amnt': 'Amount Requested', 'title': 'Loan Title', 'zip_code' : 'Zip Code', 'addr_state': 'State', 'emp_length': 'Employment Length'}

# Rename selected columns 
accepted = accepted.rename(columns=rename_dict)
```

```python
# Create a list consisting of all columns that occur in both accepted and rejected datasets
common_cols = set(accepted.columns).intersection(rejected.columns)
common_cols
```

```python
#Delete all columns that do not occur in both datasets
accepted = accepted[common_cols]
rejected = rejected[common_cols]
```

```python
accepted.describe(include='all')
```

```python
accepted = accepted.dropna()
rejected = rejected.dropna()
```

```python
rejected.describe(include='all')
```

```python
# Compare original rejection rate with rejection rate using current datasets
print(rejection_rate)
print(len(rejected) / len(accepted))
```

```python
# Calculate adjustment parameter
rej_adj = 12.230162679629018 / 21.26968154185689
rej_adj
```

```python
# Subsample to have an accepted dataset of around 35.000 loans, 
# adjust rejected dataset accordingly to match the initial rejection rate
accepted = accepted.sample(n=round(len(accepted)*0.025), random_state=888)
rejected = rejected.sample(n=round(len(rejected)*0.025*0.9*rej_adj), random_state=888)
```

```python
# Check if the rate matches the intial rejection rate
print(len(rejected) / len(accepted)*1.1)
```

```python
# Split the accepted dataset into starting dataframe and active learning dataframe
sets = train_test_split(accepted, test_size=0.1, random_state=888) 
accepted_al, accepted_start = sets
```

```python
accepted_al
```

```python
# Create the dataframe consisting of both accepted and rejected loans
all_cases_al = pd.concat([accepted_al, rejected])
```

```python
accepted_start.to_csv(f'LC_start.csv', index=False) 
accepted_al.to_csv(f'LC_accepted.csv', index=False) 
all_cases_al.to_csv(f'LC_all.csv', index=False) 
```

```python
with open('LC_start.pickle', 'wb') as f:
    pickle.dump(accepted_start, f)
    
with open('LC_accepted.pickle', 'wb') as f:
    pickle.dump(accepted_al, f)

with open('LC_all.pickle', 'wb') as f:
    pickle.dump(all_cases_al, f)

```
